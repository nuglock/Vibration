## Exemplo:
No balanceamento de dois planos de um rotor de turbina, os dados obtidos a partir da medição do desbalanceamento original, 
o peso de teste no plano direito e o peso de teste no plano esquerdo são mostrados abaixo. 

Determine o tamanho e a localização dos pesos necessários para o balanceamento dinâmico.




Os dados podem ser expressos na seguinte notação




```python
# Representação dos dados na tabela em notação complexa
import cmath         # Biblioteca matemática que define números complexos e suas operações algébricas
import math

Va = cmath.rect( 8.5 ,math.radians( 60 ))
Vb = cmath.rect( 6.5 ,math.radians( 205 ))

V_a = cmath.rect( 6.0 ,math.radians( 125 ))
V_b = cmath.rect( 4.5 ,math.radians( 230 ))

V_aa = cmath.rect( 6.0 ,math.radians( 35 ))
V_bb = cmath.rect( 10.5 ,math.radians( 160 ))

W_L = cmath.rect( 10 ,math.radians( 270 ))
W_R = cmath.rect( 12 ,math.radians( 180 ))

# Cálculo do vetores A

A_AL = (V_a - Va)/W_L
A_BL = (V_b - Vb)/W_L

A_AR = (V_aa - Va)/W_R
A_BR = (V_bb - Vb)/W_R

# Cálculo do vetores U_L r U_R

U_L = ( A_BR*Va - A_AR*Vb )/( A_BR*A_AL - A_AR*A_BL )
U_R = ( A_BL*Va - A_AL*Vb )/( A_BL*A_AR - A_AL*A_BR )

B_L = -U_L
B_R = -U_R

B_Lp = cmath.polar(B_L)
B_Rp = cmath.polar(B_R) 

print('Deve-se adicionar {:06.3f}'.format( B_Lp[0] ), 'oz em {:06.3f}'.format( 180 + math.degrees(B_Lp[1]) ), '° no plano esquerdo' )
print('Deve-se adicionar {:06.3f}'.format( B_Rp[0] ), 'oz em {:06.3f}'.format( 180 + math.degrees(B_Rp[1]) ), '° no plano direito' )

```
